# Data Summary

The diamond dataset has 53,940 use cases and 10 attributes for each diamond. The attributes data types consist of ints, floats, and objects.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = 'https://raw.githubusercontent.com/AbulHasan03/CS381_DataScience/main/diamonds.csv'
df = pd.read_csv(url)

# Shows how many diamonds are in the dataset
print("Number of use cases:", df.shape[0])

# The columns are the attributes of each diamond, the unnamed column is each diamond
print("Number of attributes:", df.shape[1])

# Display the data type measuring each attribute
print("\nData Types and Non-Null Info:")
df.info()

# Display first few rows of data
print("\n")
df.head()
```

## Exploratory Data Analysis by Attribute:

### Carat:

```python
#Carat vs Diamond
plt.hist(df['carat'], bins=30)
plt.title('Relationship between Diamonds and their Weight')
plt.xlabel('Weight (in Carats)')
plt.ylabel('Number of Diamonds')
plt.show()

print('\n')

sns.histplot(df['carat'], bins=30, kde=True)
plt.title('Relationship between Diamonds and their Weight')
plt.xlabel('Weight (in Carats)')
plt.ylabel('Number of Diamonds')
plt.show()
```

The majority of the diamonds in this dataset fall between .3 of a carat 1.3 carats. Higher carat sizes than 1.3 are rare.

### Cut:

```python
#Cut vs. Diamond
df['cut'].value_counts().plot(kind='bar')
plt.title('Cut Quality')
plt.ylabel('Number of Diamonds')
plt.show()

print('\n')

sns.countplot(x='cut', data=df)
plt.title('Cut Quality')
plt.ylabel('Number of Diamonds')
plt.show()
```

The dataset supports the idea that ideal is the most preferred cut and premium is the next best option. Very good is also a popular cut but not as high as the other two.

### Color

```python
# Diamond's Colors
df['color'].value_counts().sort_index().plot(kind='bar')
plt.title('Color Rating')
plt.xlabel('Color')
plt.ylabel('Number of Diamonds')
plt.show()

print('\n')

sns.countplot(x='color', data=df, order=sorted(df['color'].unique()))
plt.title('Color Rating')
plt.xlabel('Color')
plt.ylabel('Number of Diamonds')
plt.show()
```

G is the most frequent color along with the colors adjacent to G being high as well. This suggests that the colors are cenetered around G and the top range and worst range are less common.

### Clarity

```python
#clarity
df['clarity'].value_counts().sort_index().plot(kind='bar')
plt.title('Clarity')
plt.xlabel('Clarity')
plt.ylabel('Number of Diamonds')
plt.show()

print('\n')

sns.countplot(x='clarity', data=df, order=sorted(df['clarity'].unique()))
plt.title('Clarity')
plt.xlabel('Clarity')
plt.ylabel('Number of Diamonds')
plt.show()
```

Diamonds with a rating of SI1 and VS2 are the most common and are in the middle of the data set. The most clear, or highest rating clarity diamonds (I1 and IF) are the most rare diamonds.

### Depth

```python
#Depth and Diamonds
plt.hist(df['depth'], bins=30)
plt.title('Diamonds and their Depth')
plt.xlabel('Depth (as Percentage)')
plt.ylabel('Number of Diamonds')
plt.show()

print('\n')

sns.histplot(df['depth'], bins=30, kde=True)
plt.title('Diamonds and their Depth')
plt.xlabel('Depth (as Percentage)')
plt.ylabel('Number of Diamonds')
plt.show()
```

The percentages of the depths of the diamond are almost a normal distribution or bell curve with the data centered around the 60% mark.

### Table

```python
#Diamond Table (topmost flatness)
plt.hist(df['table'], bins=30)
plt.title('Diamonds and their Table')
plt.xlabel('Table (as Percentage)')
plt.ylabel('Number of Diamonds')
plt.show()

print('\n')

sns.histplot(df['table'], bins=30, kde=True)
plt.title('Diamonds and their Table')
plt.xlabel('Table (as Percentage)')
plt.ylabel('Number of Diamonds')
plt.show()
```

The flat top of the diamonds seem to have most of the data around the 55-60% range indicating that most diamonds have this kind of flat.

### Price:

```python
# Price vs. Count
plt.hist(df['price'], bins=30)
plt.title('Relationship between Diamonds and their Prices')
plt.xlabel('Price ($)')
plt.ylabel('Number of Diamonds')
plt.show()

print('\n')

sns.histplot(df['price'], bins=30, kde=True)
plt.title('Relationship between Diamonds and their Prices')
plt.xlabel('Price ($)')
plt.ylabel('Number of Diamonds')
plt.show()
```

The graph shows that most diamonds are priced at $5,000 or less. After that price point, the graph starts to flatten out as fewer diamonds are priced higher than that.

### X, Y, Z

```python
# XYZ Attributes
dims = ['x', 'y', 'z']
for dim in dims:
    plt.hist(df[dim], bins=30)
    plt.title(f'{dim.upper()} Distribution')
    plt.xlabel(f'{dim} (mm)')
    plt.ylabel('Number of Diamonds')
    plt.show()
    print('\n')


for dim in dims:
  sns.histplot(df['x'], bins=30, kde=True)
  plt.title(f'{dim.upper()} Distribution')
  plt.xlabel(f'{dim} (mm)')
  plt.ylabel('Number of Diamonds')
  plt.show()
  print('\n')

```

The dimensions of the diamond seem to have distributions that are all in the same 4-8mm range. This would make sense for the diamonds as it shows that their dimensions are roughly evenly split between their length, width, and depth.

# Inferences


## Proportion of diamonds between .30 and 1.08 carats

First set a condition to find the diamonds between .3 and 1.08 carats. Then find the length of those values divided by the length of all the values in the dataframe. This will give us a percentage or proportion in form (desired : total) for the diamonds.

```python
between_30_and_108 = df[(df['carat'] >= .30) & (df['carat'] <= 1.08)]
proportion_between_30_and_108 = len(between_30_and_108) / len(df)
print(f"Proportion of diamonds between .30 and 1.08 carats: {proportion_between_30_and_108:.2%}")
```

## How many of the diamonds have equal x and y dimensions?

Compare the columns for the dimensions of x and y by using a double equal sign. This will make sure the values are equal.

```python
equal_xy = df[df['x'] == df['y']]
num_equal_xy = len(equal_xy)
print(f"Number of diamonds with equal x and y dimensions: {num_equal_xy}")
```

## How many of the diamonds have a carat value less than the mean carat value?

To calculate the diamonds with carat value less than the mean; we would first calculate the mean then compare the data to the mean.

```python
mean_carat = df['carat'].mean()
less_than_mean = df[df['carat'] < mean_carat]
num_less_than_mean = len(less_than_mean)
print(f"Number of diamonds with carat value less than the mean({mean_carat:.2f}): {num_less_than_mean}")
```

## How many diamonds have a Premium cut or better? Note that the possible values for the quality of a cut are ranked in ascending order as follows: Fair / Good / Very Good / Premium / Ideal

To find the premium or better cut diamonds, we can search for the data values with a cut that is in premium and ideal since they are the top two cuts. Then we just display the length of those values.

```python
premium_or_better = df[df['cut'].isin(['Premium', 'Ideal'])]
num_premium_or_better = len(premium_or_better)
print(f"Number of diamonds with a Premium cut or better: {num_premium_or_better}")
```

## Which diamond has the highest price per carat? What is its value?

Calculate the price per carat by dividing the price by the number of carats to get the highest value per carat. Then find the location/id of the diamond with the highest price per carat.

```python
df['price_per_carat'] = df['price'] / df['carat']
highest_price_per_carat = df.loc[df['price_per_carat'].idxmax()]
print("Diamond with the highest price per carat:")
print(highest_price_per_carat)
```

##Using both Matplolib and Seaborn, make boxplots of the diamond price for each distinct cut value and discuss any conclusions you can draw from the appearance of the boxplots.

Construct a boxplot using the matplotlib and seaborn libraries. The boxplot appears to show that the less fine cuts have wider distribution of the data.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='cut', y='price', data=df)
plt.title('Diamond Price by Cut')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.show()

print('\n')

plt.figure(figsize=(10, 6))
sns.boxplot(x='cut', y='price', data=df)
plt.title('Diamond Price by Cut')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.show()
```

## Using both Matplolib and Seaborn, make a scatter plot of price vs. carat. What can we say about the relationship between those two attributes?

The scatterplot shows a correlation between the carat increasing, the price also increases. There may be other attributes affecting this price increase but the carat seems most tightly related.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='carat', y='price', data=df)
plt.title('Price vs. Carat')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()

print('\n')

plt.figure(figsize=(10, 6))
sns.scatterplot(x='carat', y='price', data=df)
plt.title('Price vs. Carat')
plt.xlabel('Carat')
plt.ylabel('Price')
```
